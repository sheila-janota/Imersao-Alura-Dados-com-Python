A seguir vamos criar um data frame do zero ficticio para criar uma media do salario de funcionarios para representar o ano com dados nulos. Aqui vou apresentar 3 formas/possibilidades de trabalhar/tratar os dados nulos.

1-Eu nao tenho, mas posso colocar/incluir alguma informacao/dados a mais na base

2- pode-se excluir esses dados que nao têm informacao (nulos) para nao poluir ou trazer lixo na minha base

3-pode-se colocar/incluir informacao a mais na nossa base baseando-se em alguma regra mais inteligente ou em uma regra menos inteligente mas que  seja mais simples. eh necessario entender o que funciona melhor para cada caso.

```python
# vou utilizar o valor da media ou mediana baseando-se em todos os valores que estao na base. isso vai ajudar a nao descalibrar/desbalancear os dados da base e eh uma estrategia comummente utilizada

#a) CRIACAO DE UM DATAFRAME FICTICIO de teste para usar de exemplo: df_salarios
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana","Bruno","Carlos","Daniele", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
    })

# --- FAZER O CALCULO DA MEDIA e MEDIANA de df_salarios

#agora vou preencher com a media do salario, ou seja tudo que for nulo vou preenchar com a media do valor.
# fillna() - preencher os valores nulos dessa base de dados df_salario na coluna salario

# df_salarios eh a minha base, salario_media eh a coluna nova que estou a criar nela e a coluna nova vai ser preenchida (fillna) com o salario.mean() que eh a media de df_salario.
# round2 = vai arredondar com 2 casa decimais. para evitar a geraçao automatica de muitas casa decimais
#deve pegar o salario varias vezes, primeiro uso a media, depois calculo e preencho a nova coluna que foi criada

# 1-CALCULO da MEDIA de df_salarios e substitui os valores nulos pela media e arrendonda o resultado para 2 casa decimais
df_salarios['salario_media']= df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))


# 2-CALCULO da MEDIANA de df_salarios e substitui os valores nulos pela mediana
#a mediana suavisa o impacto sobre valores de salario muito altos como o da Val
df_salarios['salario_mediana']= df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
# b) CRIACAO DE UM NOVO DATAFRAME FICTICIO de teste para usar de exemplo: df_temperatura
#ffill() = foward fill vai preencher/completar com o valor anterior. o valor que está para atrás
# - qual é o valor de temperatura antes de terça feira nessa caso a temperatura de segunda?R: 30° graus


df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda","Terça","Quarta","Quinta","Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
# caso o valor de temperatura de segunda estivesse com o valor nan como ficaria os valores para segunda, terça e quarta?R: podemos utilizar um novo metodo outra estratégia usando bfill()
# bfill() - back fill - vai olhar o valor que está para frente ao inve´s de pegar o valor anterior pega o valor posterior da sequencia de temperatura preenchida

df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda","Terça","Quarta","Quinta","Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
import pycountry
```

```python
# adicionar o código
# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

#Calcular média salarial por país ISO-3
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

#Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salario médio de Cientista de dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
